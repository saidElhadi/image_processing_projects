```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('wafer.jpg', cv2.IMREAD_GRAYSCALE)
img = cv2.resize(img, (500, 500))
plt.imshow(img, cmap='gray')
```

```python
edges = cv2.Canny(img, 255, 260)
plt.imshow(edges, cmap='gray')
```

```python
lines = cv2.HoughLinesP(edges, 1, np.pi/180, 270, minLineLength=400, maxLineGap=200)
# extend the lines from left to right
for line in lines:
    x1, y1, x2, y2 = line[0]
    cv2.line(img, (0,y1), (500,y2), (0,0,0), 2) 

# draw the lines
for line in lines:
    x1, y1, x2, y2 = line[0]
    cv2.line(img, (x1,y1), (x2,y2), (0,0,0), 2)
# display the result
plt.imshow(img, cmap='gray')
```

```python
# draw a horizontale line that goes through the center of the four lines drwon above without finding the edges the lines that were found before and stored in the variables lines 
# find the center of the four lines
x1, y1, x2, y2 = lines[0][0]
x3, y3, x4, y4 = lines[1][0]
x5, y5, x6, y6 = lines[2][0]
x7, y7, x8, y8 = lines[3][0]
# find the center of the four lines
x_center = int((x1+x2+x3+x4+x5+x6+x7+x8)/8)
y_center = int((y1+y2+y3+y4+y5+y6+y7+y8)/8)
# draw the line
cv2.line(img, (0,y_center), (500,y_center), (0,0,0), 2)
# display the result
plt.imshow(img, cmap='gray')
```
